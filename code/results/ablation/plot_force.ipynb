```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_interaction(dist, k, epsilon, radii = 1.0):
    """
    Given the vectors ri and rj, compute the force between them
    """

    # Combined radius of both particles (assume unit radii for now)
    #bij = 2.0                       # Ri + Rj 
    bij = radii + radii

    if dist < bij*(1 + epsilon):
        force = k*(dist - bij)
    elif dist < bij*(1 + 2*epsilon):
        force = -k*(dist - bij - 2*epsilon*bij)
    else:
        force = 0.0
    return force
```

```python
x = np.linspace(0, 6, 100)
y = [calculate_interaction(x[i], 70, 0.5) for i in range(len(x))]
```

```python
print(y)
print(np.max(y))
```

```python
plt.plot(x, y)
plt.vlines([2*(1+0.5), 2*(1+2*0.5)], -140.0 - 20, np.max(y) + 20, colors='r', linestyles='dashed')
plt.ylim([-140.0 - 20, np.max(y) + 20])
plt.xlim([0, 6])
plt.text(2*(1+0.5) - 0.6, -145 + 15, '$2 (1+\epsilon) $', ha='center', va='top', color='red')
plt.text(2*(1+2*0.5) + 0.6, -145 + 15, '$2 (1+2 \epsilon) $', ha='center', va='top', color='red')
plt.xlabel('Distance')
plt.ylabel('Norm of the interaction')
plt.show()
```

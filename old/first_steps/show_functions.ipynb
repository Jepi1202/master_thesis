```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Forces

def force(r, epsilon, R1, R2, k):
    bij = R1 + R2
    force = []
    cond1 = (1+epsilon) * bij
    cond2 = (3 * bij)

    for i in range(len(r)):
        if r[i] < cond1:
            force.append(k * (r[i]-bij))
        elif r[i] < cond2:
            force.append(-k*(r[i]-bij-2*epsilon*bij))
        else:
            force.append(0)

    return force
```

```python
r = np.linspace(0, 14, 1000)
v0 = 50 # Active force
k = 10.
epsilon = 0

R1 = 2
R2 = 2

for epsilon in [-1, -0.5, 0, 0.5, 1, 1.5, 2]:
    f = force(r, epsilon, R1,R2, k)
    plt.plot(r, f, label = f'epsilon {epsilon}')

plt.legend()
```
